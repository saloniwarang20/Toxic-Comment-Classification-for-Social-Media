```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
```

```python
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
ann_model = Sequential([
    Flatten(input_shape=(32, 32, 3)),
    Dense(512, activation='relu'),
    Dense(256, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
ann_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history_ann = ann_model.fit(x_train, y_train, epochs=10, batch_size=64, validation_data=(x_test, y_test))
```

```python
cnn_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])
```

```python
cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history_cnn = cnn_model.fit(x_train, y_train, epochs=10, batch_size=64, validation_data=(x_test, y_test))
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(history_ann.history['accuracy'], label='ANN Train Accuracy')
plt.plot(history_ann.history['val_accuracy'], label='ANN Test Accuracy')
plt.plot(history_cnn.history['accuracy'], label='CNN Train Accuracy')
plt.plot(history_cnn.history['val_accuracy'], label='CNN Test Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
```

```python
plt.subplot(1, 2, 2)
plt.plot(history_ann.history['loss'], label='ANN Train Loss')
plt.plot(history_ann.history['val_loss'], label='ANN Test Loss')
plt.plot(history_cnn.history['loss'], label='CNN Train Loss')
plt.plot(history_cnn.history['val_loss'], label='CNN Test Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
ann_loss, ann_acc = ann_model.evaluate(x_test, y_test)
print(f'ANN Test Accuracy: {ann_acc:.4f}, Loss: {ann_loss:.4f}')
```

```python
cnn_loss, cnn_acc = cnn_model.evaluate(x_test, y_test)
print(f'CNN Test Accuracy: {cnn_acc:.4f}, Loss: {cnn_loss:.4f}')
```
